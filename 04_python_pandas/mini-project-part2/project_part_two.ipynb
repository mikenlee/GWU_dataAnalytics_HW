```python
# Dependencies
import pandas as pd
```

```python
# load CSV
bc_survey_csv = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory = False)
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
survey_df = bc_survey_csv.iloc[:, [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
```

```python
#find the columns that use 0 and 1
for col in survey_df:
    print(col, survey_df[col].unique())

# Change reading 0 and 1 to No and Yes, respectively
response_cols = ["AttendedBootcamp", "BootcampFinish", "BootcampFullJobAfter", "BootcampLoanYesNo", "BootcampRecommend", "HasChildren" ]
survey_df[response_cols] = survey_df[response_cols].replace({0.0 : "No", 1.0 : "Yes"}) 
```

```python
# Extract rows for only those who attended a bootcamp
attended_df = survey_df.loc[survey_df["AttendedBootcamp"] == "Yes"]
attended_df
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
bootcamp_df = attended_df["BootcampName"].value_counts().rename_axis("BootcampName").reset_index(name = "Attendees")
bootcamp_df
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
recommend_df = attended_df.loc[attended_df["BootcampRecommend"] == "Yes", :]["BootcampName"].value_counts()
recommend_df = recommend_df.rename_axis("BootcampName").reset_index(name = "Recommendation Count")
recommend_df
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merge_df = pd.merge(bootcamp_df, recommend_df, on = "BootcampName", how = "left")
merge_df
```

```python
#defining a function to convert value into percent format
def as_percent(amount):
    return "{0:.2f}%".format(amount * 100)

# Calculate percentage of eac bootcamp's students who are recommenders
merge_df["% Recommended"] = merge_df["Recommendation Count"] / merge_df["Attendees"]
merge_df = merge_df.sort_values("% Recommended", ascending = False)
merge_df["% Recommended"] = merge_df["% Recommended"].map(as_percent)
merge_df
```

```python
# Export to excel and remove index
merge_df.to_excel("mini_proj2_output.xlsx", index = False, header = True)
```

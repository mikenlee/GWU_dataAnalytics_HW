```python
# Dependencies
import pandas as pd
```

```python
# load CSV
survey_csv = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Read with pandas
```

```python
# Inspect all columns
survey_csv.dtypes
# survey_csv.head()
survey_csv.describe()

# Remove the rows with missing data
survey_csv = survey_csv.dropna(how="all")
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
survey_df = survey_csv.iloc[:, [0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
# survey_df["AttendedBootcamp"].dtypes
# survey_df["AttendedBootcamp"].value_counts()
# survey_df["AttendedBootcamp"].unique()
# for col in survey_df:
#     print(col, survey_df[col].unique())
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
response_cols = ["AttendedBootcamp", "BootcampFinish", "BootcampFullJobAfter", "BootcampLoanYesNo", "BootcampRecommend", "HasChildren" ]
survey_df[response_cols] = survey_df[response_cols].replace({0.0 : "No", 1.0 : "Yes"}) 
```

```python
# test the response replacement
# survey_df.loc[survey_df["AttendedBootcamp"] == "Yes"]
# survey_df.loc[survey_df["AttendedBootcamp"].isna() == True, :]
```

```python
# Calculate total number of respondents in survey
total_respondents = len(survey_df.index)
total_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attended_df = survey_df.loc[survey_df["AttendedBootcamp"] == "Yes"]
attended_df
```

```python
# Calculate average age of attendees
avg_age = attended_df["Age"].mean()
# attended_df["Age"].max()

# Calculate how many people attended a bootcamp
total_attendees = attended_df["AttendedBootcamp"].count()

# Calculate how many attendees hold degrees. (Assumed associates and higher)
degrees = ["bachelor's degree", "master's degree (non-professional)",
       'professional degree (MBA, MD, JD, etc.)', "associate's degree", 'Ph.D.']
degree_attendees = attended_df["SchoolDegree"].isin(degrees).sum()

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
attended_df["Gender"].value_counts()
male_attendees = attended_df["Gender"].loc[attended_df["Gender"] == "male"].count()
female_attendees = attended_df["Gender"].loc[attended_df["Gender"] == "female"].count()

self_id = ["genderqueer", "trans", "agender"]
selfid_attendees = attended_df["Gender"].isin(self_id).sum()
total_gender_attendees = attended_df["Gender"].notna().sum()

# Calculate percentage of respondents who attended a bootcamp
attendees_perc = total_attendees / total_respondents

# Calculate percentage of respondents belonging to each gender
male_perc = male_attendees / total_gender_attendees
female_perc = female_attendees / total_gender_attendees
self_perc = selfid_attendees / total_gender_attendees

# Calculate percentage of attendees with a school degree
degree_perc = degree_attendees / total_attendees

# Calculate average post-bootcamp salary
postboot_salary = attended_df["BootcampPostSalary"].mean()
male_attendees
```

```python
#defining a function to convert value into currency format
def as_currency(amount):
        return '${:,.2f}'.format(amount)

def as_percent(amount):
    return "{0:.2f}%".format(amount * 100)
    
# Create a new table consolodating above calculations
bootcamp_dict = {"Total Surveyed" : total_respondents,
"Total Bootcamp Attendees" : total_attendees,
"% Attended Bootcamp" : as_percent(attendees_perc),
"AVG Age" : round(avg_age, 2),
"Has a degree" : as_percent(degree_perc), #assumed associates degree and higher
"% Male" : as_percent(male_perc),
"% Female" : as_percent(female_perc),
"% Non Gender Specific" : as_percent(self_perc),
"AVG Post Bootcamp Salary" : as_currency(postboot_salary)}

output_df = pd.DataFrame(bootcamp_dict, index = ['i',])
output_df
```
